```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from heapq import nlargest

def nltk_summarize(text, num_sentences=1):
    # Tokenize the text into sentences
    sentences = sent_tokenize(text)

    # Tokenize the sentences into words and remove stopwords
    stop_words = set(stopwords.words("english"))
    words = word_tokenize(text)
    words = [word.lower() for word in words if word.isalnum()]
    words = [word for word in words if word not in stop_words]

    # Calculate word frequencies
    word_freq = nltk.FreqDist(words)

    # Calculate sentence scores based on word frequencies
    sentence_scores = {}
    for sentence in sentences:
        for word in nltk.word_tokenize(sentence.lower()):
            if word in word_freq:
                if len(sentence.split(" ")) < 30:  # Filter out long sentences
                    if sentence not in sentence_scores:
                        sentence_scores[sentence] = word_freq[word]
                    else:
                        sentence_scores[sentence] += word_freq[word]

    # Get the top N sentences with highest scores
    summarized_sentences = nlargest(num_sentences, sentence_scores, key=sentence_scores.get)

    return ' '.join(summarized_sentences)

df['summary_text'] = df['web_text'].apply(nltk_summarize)
```

```python
def preprocesstext(text):
    text = text.lower() #convert to lowercase
    text = re.sub(r'[^a-zA-Z]', ' ', text) #remove symbols
    tokens = word_tokenize(text)
    stopword = nltk.corpus.stopwords.words('english')
    tokens = [token for token in tokens if token not in stopword]
    lemmatizer = nltk.WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(tokens)
```

```python
def get_sentiment_score(text):
    if isinstance(text, float):
        return None
    else:
    # Initialize the sentiment analyzer
        sid = SentimentIntensityAnalyzer()
        
        # Get the sentiment scores
        sentiment_scores = sid.polarity_scores(text)
        
        # Return the compound sentiment score
        return sentiment_scores['compound']
```

```python
def extract_entities(text):
    words = word_tokenize(text)
    pos_tags = pos_tag(words)
    ner_result = ne_chunk(pos_tags)

    entities = []
    for chunk in ner_result:
        if hasattr(chunk, 'label') and chunk.label:
            entity = ' '.join(c[0] for c in chunk.leaves())
            entities.append(entity)

    return entities
```

```python
df['nltk_entities'] = df['web_text'].apply(extract_entities)
```

```python
import gensim
from gensim import corpora
from gensim.models import LdaModel
from gensim.utils import simple_preprocess
from gensim.models.phrases import Phrases, Phraser
```

```python
def preprocess(tokens):
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    return [lemmatizer.lemmatize(stemmer.stem(token)) for token in tokens if token not in stop_words]
```

```python
df['preprocessed_tokens'] = df['tokenized_text'].apply(preprocess)
```

```python
bigram = Phrases(df['preprocessed_tokens'], min_count=5, threshold=100)
trigram = Phrases(bigram[df['preprocessed_tokens']], min_count=5, threshold=100)
```

```python
df['bigram_tokens'] = list(bigram[df['preprocessed_tokens']])
df['trigram_tokens'] = list(trigram[bigram[df['preprocessed_tokens']]])
```

```python
df['combined_tokens'] = df['preprocessed_tokens'] + df['bigram_tokens'] + df['trigram_tokens']
```

```python
dictionary = corpora.Dictionary(df['combined_tokens'])
corpus = [dictionary.doc2bow(doc) for doc in df['combined_tokens']]
```

```python
num_topics = 50  # You can change this to the number of topics you want
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print(f'Topic: {idx}')
    print(f'Top words: {topic}\n')
```

```python
def get_dominant_topic_id(lda_model, document):
    topic_distribution = lda_model[document]
    dominant_topic = max(topic_distribution, key=lambda x: x[1])[0]
    return dominant_topic

# Get the dominant topic for each row in the DataFrame
df['dominant_topic'] = df['combined_tokens'].apply(lambda x: get_dominant_topic_id(lda_model, dictionary.doc2bow(x)))
```

```python
def get_topic_words(lda_model, num_words=20):
    topic_words = {}
    for topic_id in range(num_topics):
        words = [word for word, _ in lda_model.show_topic(topic_id, topn=num_words)]
        topic_words[topic_id] = words
    return topic_words
topics_df = pd.DataFrame(get_topic_words(lda_model, num_words=20))
```

```python
# Rename columns to represent the topic number
topics_df.columns = [str(topic_id) for topic_id in topics_df.columns]

# Export the topics and their top words to an Excel file
topics_df.to_excel('Dataset/topics_and_top_words.xlsx', index=False)
```
